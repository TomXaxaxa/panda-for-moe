```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
result_csv_files = [
    '../eval_results/test_base_metrics.csv',
    '../eval_results/test_20000_metrics.csv',
    '../eval_results/test_40000_metrics.csv',
    '../eval_results/test_60000_metrics.csv',
    '../eval_results/test_80000_metrics.csv',
    '../eval_results/test_100000_metrics.csv',
]
```

```python
from collections import defaultdict

smape_dict = defaultdict(list)
for file in result_csv_files:
    print('Processing file:', file)
    # read the CSV file
    df = pd.read_csv(file)
    # get the smape column
    smape = df['sMAPE']
    dyst = df['dataset']
    for i in range(len(smape)):
        smape_dict[dyst[i]].append(smape[i])
```

```python
print(smape_dict)
```

```python
list(smape_dict.keys())
```

```python
# plot smape_dict
fig, ax = plt.subplots(figsize=(5, 5))
for key in list(smape_dict.keys())[:10]:
    value = smape_dict[key]
    ax.plot(value, label=key)
    ax.set_yscale('log')
    ax.set_xlabel('Iterations')
ax.set_ylabel('sMAPE')
ax.set_title('sMAPE')
ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.show()
```
